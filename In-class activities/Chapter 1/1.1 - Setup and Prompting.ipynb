## Prompting practice
Suppose you wanted to make a tic-tac-toe game. You want to use Cursor's Agent (or another LLM) to help you create it. Discuss the following questions:

1. What information should you be giving the LLM? Is it enough to say "make tic tac toe"? If not, what else would you include? Make a comprehensive list.
2. You should always start complex projects like this with a **planning prompt.** This is a prompt telling the LLM that you are just planning right now, you don't want to actually write any code yet. Write out your planning prompt.
3. After it writes the plan, look over it! Are you happy with it? Are there changes that should be made? Did you think of a new feature you want to implement? Don't be afraid to go back and forth with your LLM to refine your plan. 
4. Once you have a plan, give it back to the LLM, and ask it to create the steps that it will go through in creating this game. Again, _we don't want it to write any code yet!_ All we want is a detailed plan.
5. Once your plan is complete, ask your LLM to do the first step (from the steps determined in what you just did). Test it out! Does it run? Does anything work? Be as critical as possible to make sure things are how you want. Once you've got that completed, move on to the next step.
